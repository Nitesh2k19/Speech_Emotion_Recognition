load audio file


```python
import librosa
audio_path = '/content/trumpet_128kbps.mp3'
x,sr = librosa.load(audio_path)
```

listen audio file

```python
import IPython.display as ipd
ipd.Audio(audio_path)
```

plot the wave form

```python
%matplotlib inline
import sklearn
import matplotlib.pyplot as plt
import librosa.display
plt.figure()
librosa.display.waveplot(x,sr)
```

plot the spectograph

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure()
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

feature extraction


```python
sample_duration = 1/sr
signal_duration = sample_duration*len(x)
print(sample_duration)
print(signal_duration)
```
